## Percabangan (If / Elif / Else)
Percabangan membantu program mengambil keputusan berdasarkan kondisi tertentu.  
Contoh: memberi grade berdasarkan nilai ujian, atau menentukan genap/ganjil.

```python
# Contoh genap/ganjil
n = 110
if n % 2 == 0:
    print(n, "adalah GENAP")
else:
    print(n, "adalah GANJIL")
```

```python
# Contoh grading
nilai = 78
if nilai >= 85:
    grade = "A"
elif nilai >= 70:
    grade = "B"
elif nilai >= 55:
    grade = "C"
else:
    grade = "D"
print("Grade:", grade)
```

## Pengulangan For
`for` digunakan untuk mengulang item dalam urutan (list, range, dll).  
`range(n)` menghasilkan 0..n-1; `enumerate(seq)` memberi pasangan (index, nilai).

```python
buah = ["apel", "pisang", "jeruk"]
for i, b in enumerate(buah, start=1):
    print(f"{i}. {b}")
```

```python
total = 0
for i in range(1, 6):  # 1..5
  print(i)
print("Jumlah 1..5 =", total)
```

## Perulangan While
`while` berjalan selama kondisi bernilai True.  
Pastikan kondisi akan berubah agar tidak terjadi infinite loop.  
Kita juga bisa gunakan `break` (berhenti) dan `continue` (lewati iterasi).

```python
i = 1
while i <= 5:       # selama i masih <= 5, lakukan isi loop
    print(i)        # tampilkan i
    i += 1          # naikkan i agar suatu saat kondisi jadi False / increment i = i + 1
```

```python
i = 1
while i <= 5:
    print(i)
    if i == 2:
        break      # keluar dari loop sekarang juga
    i += 1
```

```python
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue   # lompat ke iterasi berikutnya, baris di bawahnya dilewati
    print(i)
```

## Fungsi (def, parameter, return, docstring)
Fungsi mengemas logika agar mudah dipakai ulang.  
- Didefinisikan dengan `def`, menerima parameter, dan mengembalikan nilai dengan `return`.  
- Gunakan docstring `"""..."""` untuk menjelaskan fungsi dan parameternya.

```python
# mendefinisikan fungsi hitung_bmi
def hitung_bmi(berat_kg: float, tinggi_m: float) -> float:
    """Menghitung BMI = berat / (tinggi^2)."""
    bmi = berat_kg / (tinggi_m ** 2)
    return bmi
```

```python
hitung_bmi(80, 1.75) # pemanggilan fungsi hitung_bmi
```

```python
def kategori_bmi(bmi: float) -> str:
    """Mengembalikan kategori BMI berdasarkan nilai BMI."""
    if bmi < 18.5:
        return "Kurus"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Berlebih"
    else:
        return "Obesitas"
```

```python
bmi = hitung_bmi(90, 1.75)
print("BMI:", round(bmi, 2), "| Kategori:", kategori_bmi(bmi))
```

## Mini-Project: Kasir Mini
Kita gabungkan konsep yang sudah dipelajari:
- **Dict** sebagai daftar harga.
- **Loop** untuk memproses pesanan.
- **If/Else** untuk menghitung diskon.
- **Fungsi** agar perhitungan rapi dan dapat dipakai ulang.
Hasil akhirnya adalah struk sederhana di terminal.

```python
# Daftar harga (IDR)
harga = {
    "apel": 12000,
    "pisang": 8000,
    "jeruk": 10000,
    "mangga": 15000
}
```

```python
def hitung_diskon(total):
    if total >= 100_000:
        return 0.10 * total
    elif total >= 50_000:
        return 0.05 * total
    return 0.0
```

```python
keranjang = {}  # item -> qty

# Simulasi input (biar langsung jalan untuk pemula)
pesanan = [("apel", 2), ("mangga", 3), ("pisang", 1), ("apel", 2)]

for item, qty in pesanan:
    if item in harga and qty > 0:
        keranjang[item] = keranjang.get(item, 0) + qty

    print("pesanan", item,"jumlahnya", qty, "buah")
```

```python
keranjang.items()
```

```python
# Hitung total
subtotal = sum(harga[i] * q for i, q in keranjang.items())
diskon = hitung_diskon(subtotal)
total = subtotal - diskon
```

```python
# Cetak struk
print("=== STRUK MINI ===")
for i, q in keranjang.items():
    print(f"{i:<7} x{q} @ {harga[i]:>6,} = {harga[i]*q:>8,}")
print("-" * 26)
print(f"Subtotal: {subtotal:>12,}")
print(f"Diskon  : {diskon:>12,}")
print(f"TOTAL   : {total:>12,}")
```

## Bonus (Opsional): Error Handling & Comprehension
- **Error handling** (`try/except`) membuat program tetap berjalan meski ada input tidak valid.
- **List/Dict Comprehension** adalah cara singkat membangun koleksi dari iterasi.


```python
# try/except untuk input campuran
texts = ["10", "lima", "7.5"]
angka = []
for t in texts:
    try:
        angka.append(float(t))
    except ValueError:
        print("Lewati input tidak valid:", t)
print("Angka valid:", angka)
```

```python
# Comprehension: kuadrat 1..5, lalu dict item->total_harga
kuadrat = [i*i for i in range(1, 6)]
print("Kuadrat:", kuadrat)

keranjang = {"apel": 2, "jeruk": 1}
harga = {"apel": 12000, "jeruk": 10000}
total_per_item = {item: qty * harga[item] for item, qty in keranjang.items()}
print("Total per item:", total_per_item)
```
