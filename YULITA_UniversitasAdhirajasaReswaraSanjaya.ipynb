```python
#TUGAS 1

Nama_produk_pertama= "Kopi Pagi"
harga_produk_pertama=18000.5
Nama_produk_kedua= "Roti Cokelat"
harga_produk_kedua=10000
status_ketersediaan_roti=True #roti tersedia

print("Nama produk 1:", Nama_produk_pertama)
print("Harga produk 1:",harga_produk_pertama)
print("Nama produk 2:",Nama_produk_kedua)
print("Harga produk 2:",harga_produk_kedua)
print("Status ketersediaan Roti:",status_ketersediaan_roti)
```

```python
#TUGAS 2

#minta input pesanan pengguna
jumlah_kopi_str=input("Masukan jumlah pesanan Kopi:")
jumlah_roti_str=input("Masukan jumlah pesanan roti:")

#cek tipe data dari masing masing variabel
print("Tipe data awal jumlah  Kopi:",type(jumlah_kopi_str))
print("Tipe data awal jumlah  roti:",type(jumlah_roti_str))

#konversi tipe data string --> int
jumlah_kopi_int=int(jumlah_kopi_str)
jumlah_roti_int=int(jumlah_roti_str)

#tipe data setelah di konversi
print("Tipe data setelah konversi:",type(jumlah_kopi_int))
print("Tipe data setelah konversi:",type(jumlah_roti_int))
```

```python
#TUGAS 3

# data harga (dari Tugas 1)
harga_produk_pertama=18000.5
harga_produk_kedua=10000

total_harga_kopi=harga_produk_pertama * jumlah_kopi_int
total_harga_roti=harga_produk_kedua * jumlah_roti_int

print("Total harga kopi:",total_harga_kopi)
print("Total harga roti:",total_harga_roti)

#menghitung total keseluruhan
total_belanja=total_harga_kopi + total_harga_roti
print("Total belanja keseluruhan:",total_belanja)

#menghitung kembalian
uang_bayar=50000
print("Uang yang dibayarkan:",uang_bayar)

kembalian=uang_bayar - total_belanja
print("kembalian:",kembalian)
```
