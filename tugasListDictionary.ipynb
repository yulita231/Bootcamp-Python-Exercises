```python
#Buat list nama siswa
nama_siswa = ["Yulita", "Resa", "Fahmi", "Kaka", "indriyani"]
print("1. List awal nama siswa:")
print(nama_siswa)
print("-" * 30)

# Langkah 2: Buat dictionary nilai siswa
nilai_siswa = {
    "Yulita": 85,
    "Resa": 92,
    "Fahmi": 78,
    "Kaka": 88,
    "indriyani": 95
}
print("2. Dictionary awal nilai siswa:")
print(nilai_siswa)
print("-" * 30)

print("\n--- Manipulasi List ---")
# menambahkan satu nama siswa baru ke akhir list
nama_siswa.append("sestri")
print("Menambahkan 'sestri':", nama_siswa)

# menghapus satu siswa tertentu dari list
nama_siswa.remove("Fahmi")
print("Menghapus 'Fahmi':", nama_siswa)

# mencetak nama siswa pertama dan terakhir
print("Nama siswa pertama:", nama_siswa[0])
print("Nama siswa terakhir:", nama_siswa[-1])

# mencetak 3 siswa terakhir menggunakan slicing
print("Tiga siswa terakhir:", nama_siswa[-3:])

#  enumerate() untuk mencetak index dan nama siswa
print("\nDaftar siswa dengan enumerate():")
for index, nama in enumerate(nama_siswa):
    print(f"Index: {index}, Nama: {nama}")

# --- Manipulasi Dictionary ---
print("\n--- Manipulasi Dictionary ---")
# mengubah nilai ujian salah satu siswa
nilai_siswa["Yulita"] = 90
print("Mengubah nilai Budi menjadi 90:", nilai_siswa)

# menghapus data nilai siswa lain menggunakan pop()
nilai_dihapus = nilai_siswa.pop("Kaka")
print(f"Menghapus data 'kaka' (nilai: {nilai_dihapus}).")
print("Dictionary setelah menghapus 'kaka':", nilai_siswa)

# mengganti key nama siswa tertentu menjadi nama baru
# membuat item baru dengan key baru, lalu hapus item dengan key lama
nilai_siswa["Chandra"] = nilai_siswa.pop("Resa")
print("Mengganti key 'Resa' menjadi 'Chandra':")
print("Dictionary hasil akhir:", nilai_siswa)
print("-" * 30)
```
