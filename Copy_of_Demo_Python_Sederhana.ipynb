# 3. Operasi Dasar

```python
a = 10
b = 3

print(a + b)  # Penjumlahan
print(a - b)  # Pengurangan
print(a * b)  # Perkalian
print(a / b)  # Pembagian
print(a // b) # Pembagian bulat
print(a % b)  # Modulo
```

# 4. Struktur control

## Percabangan

```python
x = 10
if x > 5:
    print("x lebih besar dari 5")
elif x == 5:
    print("x sama dengan 5")
else:
    print("x kurang dari 5")
```

## Perulangan

```python
# For loop
for i in range(5):
    print(i)

# While loop
count = 0
while count < 5:
    print(count)
    count += 1
```
