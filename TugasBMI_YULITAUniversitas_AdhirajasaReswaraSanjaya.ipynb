```python
# Menerima input berat dan tinggi dari pengguna
# Gunakan float() agar input bisa berupa angka desimal
try:
    berat = float(input("Masukkan berat badan Anda (kg): "))

    # Menerima input tinggi dalam cm
    tinggi_cm = float(input("Masukkan tinggi badan Anda (cm): "))

    # Mengubah tinggi dari cm ke m untuk perhitungan BMI
    tinggi_m = tinggi_cm / 100

    # Menghitung BMI
    bmi = berat / (tinggi_m ** 2)

    # Menampilkan hasil BMI dengan 2 angka di belakang koma
    print(f"\nBMI Anda adalah: {bmi:.2f}")

    # Menentukan kategori BMI dengan if, elif, else
    if bmi < 18.5:
        print("Kategori: Underweight (Kurus)")
        # Nested condition untuk memberikan pesan tambahan
        if bmi < 16.0:
            print("Anda berada dalam kategori Underweight yang parah.")
        else:
            print("Anda berada dalam kategori Underweight ringan.")

    elif 18.5 <= bmi < 25:
        # Menggunakan logical operator '<' dan '>='
        print("Kategori: Normal")

    elif 25 <= bmi < 30:
        print("Kategori: Overweight (Gemuk)")
        # Contoh nested condition lainnya
        if bmi >= 25 and bmi < 27:
            print("Kategori: Overweight (Risiko)")
        else:
            print("Kategori: Overweight (Penting untuk diperhatikan)")

    else: # Jika BMI >= 30
        print("Kategori: Obese (Obesitas)")
        if bmi >= 30 and bmi < 35:
            print("Kategori: Obesitas Kelas I")
        elif bmi >= 35 and bmi < 40:
            print("Kategori: Obesitas Kelas II")
        else:
            print("Kategori: Obesitas Kelas III (Morbid)")

except ValueError:
    print("Input yang Anda masukkan tidak valid. Harap masukkan angka.")
except ZeroDivisionError:
    print("Tinggi badan tidak boleh nol.")
```
